```python
class Problem:
    def initial_state(self):
        # Define the initial state
        return 'A'

    def goal_test(self, state):
        # Define the goal test
        return state == 'G'

    def expand(self, state):
        # Define how to expand a state
        # Example graph structure as an adjacency list
        graph = {
            'A': ['B', 'C'],
            'B': ['D', 'E'],
            'C': ['F'],
            'D': [],
            'E': ['G'],
            'F': [],
            'G': []
        }
        return graph[state]

    def solution(self, state):
        # Define how to construct a solution
        return f'Solution found: {state}'

def graph_search(problem):
    # Initialize the frontier with the initial state of the problem
    frontier = [problem.initial_state()]
    # Initialize the explored set to be empty
    explored = set()

    while True:
        # If the frontier is empty, return failure
        if not frontier:
            return 'failure'

        # Choose a leaf node and remove it from the frontier
        node = frontier.pop(0)

        # If the node contains a goal state, return the corresponding solution
        if problem.goal_test(node):
            return problem.solution(node)

        # Add the node to the explored set
        explored.add(node)

        # Expand the chosen node, adding the resulting nodes to the frontier
        for child in problem.expand(node):
            if child not in frontier and child not in explored:
                frontier.append(child)

# Example usage:
problem = Problem()
result = graph_search(problem)
print(result)
```

```python
class Human:

    # A class attribute. It is shared by all instances of this class
    species = "H. sapiens"

    # Basic initializer, this is called when this class is instantiated.
    # Note that the double leading and trailing underscores denote objects
    # or attributes that are used by Python but that live in user-controlled
    # namespaces. Methods(or objects or attributes) like: __init__, __str__,
    # __repr__ etc. are called special methods (or sometimes called dunder
    # methods). You should not invent such names on your own.
    def __init__(self, name):
        # Assign the argument to the instance's name attribute
        self.name = name

        # Initialize property
        self._age = 0   # the leading underscore indicates the "age" property is
                        # intended to be used internally
                        # do not rely on this to be enforced: it's a hint to other devs

    # An instance method. All methods take "self" as the first argument
    def say(self, msg):
        print("{name}: {message}".format(name=self.name, message=msg))

    # Another instance method
    def sing(self):
        return 'yo... yo... microphone check... one two... one two...'

    # A class method is shared among all instances
    # They are called with the calling class as the first argument
    @classmethod
    def get_species(cls):
        return cls.species

    # A static method is called without a class or instance reference
    @staticmethod
    def grunt():
        return "*grunt*"

    # A property is just like a getter.
    # It turns the method age() into a read-only attribute of the same name.
    # There's no need to write trivial getters and setters in Python, though.
    @property
    def age(self):
        return self._age

    # This allows the property to be set
    @age.setter
    def age(self, age):
        self._age = age

    # This allows the property to be deleted
    @age.deleter
    def age(self):
        del self._age


# When a Python interpreter reads a source file it executes all its code.
# This __name__ check makes sure this code block is only executed when this
# module is the main program.
if __name__ == '__main__':
    # Instantiate a class
    i = Human(name="Ian")
    print(i.say("hi"))                     # "Ian: hi"
    j = Human("Joel")
    j.say("hello")                  # "Joel: hello"
    # i and j are instances of type Human; i.e., they are Human objects.

    # Call our class method
    i.say(i.get_species())          # "Ian: H. sapiens"
    # Change the shared attribute
    Human.species = "H. neanderthalensis"
    i.say(i.get_species())          # => "Ian: H. neanderthalensis"
    j.say(j.get_species())          # => "Joel: H. neanderthalensis"

    # Call the static method
    print(Human.grunt())            # => "*grunt*"

    # Static methods can be called by instances too
    print(i.grunt())                # => "*grunt*"

    # Update the property for this instance
    i.age = 42
    # Get the property
    i.say(i.age)                    # => "Ian: 42"
    j.say(j.age)                    # => "Joel: 0"
    # Delete the property
    del i.age
    # i.age                         # => this would raise an AttributeError
```

```python
import math

# Define the graph (adjacency list)
graph = {
    "Oradea": {"Zerind": 71, "Sibiu": 151},
    "Zerind": {"Oradea": 71, "Arad": 75},
    "Arad": {"Zerind": 75, "Sibiu": 140, "Timisoara": 118},
    # ... (add other cities and their connections)
    "Fagaras": {"Sibiu": 99, "Bucharest": 211},
    # ...
}

# Heuristic function (straight-line distance to Bucharest)
city_coordinates = {
    "Oradea": (291, 152),
    "Zerind": (233, 160),
    "Arad": (210, 213),
    # Define coordinates for other cities
    "Fagaras": (345, 116),
    "Bucharest": (366, 0)
    # ...
}

def heuristic(city):
    x1, y1 = city_coordinates[city]
    x2, y2 = city_coordinates["Bucharest"]
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def a_star_search(start, goal):
    open_list = [(start, 0)]
    closed_list = set()
    g_values = {start: 0}
    parents = {}

    while open_list:
        current, g = open_list.pop(0)
        if current == goal:
            # Reconstruct path and return it
            path = []
            while current != start:
                path.append(current)
                current = parents[current]
            path.append(start)
            return path[::-1]  # Reverse the path
        closed_list.add(current)

        for neighbor, cost in graph.get(current, {}).items():
            if neighbor in closed_list:
                continue
            neighbor_g = g + cost
            if neighbor not in g_values or neighbor_g < g_values[neighbor]:
                h = heuristic(neighbor)
                f = neighbor_g + h
                open_list.append((neighbor, f))
                g_values[neighbor] = neighbor_g
                parents[neighbor] = current
        open_list.sort(key=lambda x: x[1])

    return "Goal unreachable"

# Example usage
start_city = "Oradea"
goal_city = "Fagaras"
result = a_star_search(start_city, goal_city)
print("Path from", start_city, "to", goal_city, ":", result)
```

```python
import heapq

# Define the graph with distances between nodes
graph = {
    'Arad': [('Zerind', 75), ('Timisoara', 118), ('Sibiu', 140)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Rimnicu Vilcea', 80), ('Fagaras', 99)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90)],
    'Giurgiu': [('Bucharest', 90)]
}

# Heuristic values (straight-line distance to Bucharest)
heuristics = {
    'Arad': 366,
    'Zerind': 374,
    'Oradea': 380,
    'Timisoara': 329,
    'Lugoj': 244,
    'Mehadia': 241,
    'Drobeta': 242,
    'Craiova': 160,
    'Sibiu': 253,
    'Rimnicu Vilcea': 193,
    'Fagaras': 176,
    'Pitesti': 100,
    'Bucharest': 0,
    'Giurgiu': 77
}

def a_star(graph, heuristics, start, goal):
    # Priority queue to store (cost, current_node, path)
    open_list = [(0 + heuristics[start], 0, start, [start])]
    heapq.heapify(open_list)
    closed_list = set()

    while open_list:
        # Get the node with the lowest cost
        _, cost, current_node, path = heapq.heappop(open_list)

        if current_node in closed_list:
            continue

        # Goal check
        if current_node == goal:
            return path, cost

        closed_list.add(current_node)

        for neighbor, weight in graph.get(current_node, []):
            if neighbor in closed_list:
                continue

            new_cost = cost + weight
            heapq.heappush(open_list, (new_cost + heuristics[neighbor], new_cost, neighbor, path + [neighbor]))

    return None, float('inf')

# Run A* algorithm from Arad to Bucharest
path, cost = a_star(graph, heuristics, 'Arad', 'Bucharest')
print(f"Path: {path}")
print(f"Cost: {cost}")
```

```python
#question 3
def is_prime(n):
  if n<1:
    return False
  for i in range(2,n):
    if n% i==0:
      return False
  return True
print (is_prime(2))
print (is_prime(10))
print (is_prime(5))
```

```python
#question num 7
def fibonacci_sequence(max_num):
  a,b=0,1
  result=f"{a}+{b}+"
  total=a+b
  while total <=max_num:
    a,b=b,a+b
    total +=b
    result=result+f" {b}+"
  return result[:-1]
  result=result+f"={total}"
  return result
print(fibonacci_sequence(20))

```

```python
#Question num 8
def simple_interest(p,r,t):
  return p*r*t/100
principle=float (input("enter the principle amount:"))
time=float (input("enter the time period (in years):"))
rate=float (input("enter the rate of the interest(in percentage):"))
print("simple interest ;",simple_interest(principle,time,rate))
```

```python
#question num 9
import math
def lcm(x,y):
  gcd=math.gcd(x,y)
  return(x*y)//gcd
print(lcm(12,15))


```

```python
#question num 10
import math
class Shape:
  def area(self):
    pass
class Rectangle(Shape):
  def __init__(self,lenght,width):
    self.length=lenght
    self.width=width
  def area(self):
    return self.length*self.width
class Circle(Shape):
  def __init__(self,radius):
    self.radius=radius
  def area(self):
   return math.pi*self.radius**2
class Triangle(Shape):
  def __init__(self,base,height):
    self.base=base
    self.height=height
  def area(self):
    return 0.5*self.base*self.height
class Square(Rectangle):
  def __init__(self,side):
    super().__init__(side,side)
rectangle=Rectangle(5,4)
print("Area of rectangle:",rectangle.area())
circle =Circle(3)
print("Area of the circle:",circle.area())
triangle=Triangle(6,8)
print("Area of the triangle :",triangle.area())
square=Square(5)
print("Area of the square :",square.area())
```
